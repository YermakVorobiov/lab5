```python
import matplotlib.pyplot as plt
import numpy as np


x1 = np.arange(-5, 5, 1)
x2 = (73+10.2*x1)/9.3
x3 = (173-5.3*x1)/15.8
x4 = (159-17.2*x1)/10.3
A = [[-10.2, 9.3], [5.3, 15.8]]
A1 = [[5.3, 15.8], [17.2, 10.3]]
A2 = [[17.2, 10.3], [-10.2, 9.3]]
B = [[73], [173]]
B1 = [[173], [159]]
B2 = [[159], [73]]
fig, ax = plt.subplots(figsize=(5, 5), dpi = 150)
plt.grid()
f0 = plt.plot(x1, x2, color = 'purple', linestyle='dashed');
plt.text(-2, 4.5, '$-10.2x_1+9.3x_2=73$', fontsize = 6, rotation=25, color='purple')
f1 = plt.plot(x1, x3, color = 'red', linestyle='dotted');
plt.text(-2, 10, '$5.3x_1+15.8x_2=173$', fontsize = 6, rotation=-7, color='red')
f2 = plt.plot(x1, x4, color = '#149fff', linestyle='dashdot');
plt.text(-4, 18, '$17.2x_1+10.3x_2=159$', fontsize = 6, rotation=-35, color='#149fff')
plt.legend(['f[0]', 'f[1]', 'f[2]'])
idx = np.argwhere(np.diff(np.sign(x2 - x3))).flatten()
f0f1 = np.linalg.solve(A, B)
f1f2 = np.linalg.solve(A1, B1)
f2f0 = np.linalg.solve(A2, B2)
plt.plot(f0f1[0], f0f1[1], color = '#ff14f8', marker='.')
plt.annotate('The intersection point between f[0] and f[1]', fontsize = 7, xy=(f0f1[0], f0f1[1]), xytext=(f0f1[0]-3.5, f0f1[1]+8),
            arrowprops=dict(width = 0.5, headlength = 5, headwidth = 3, facecolor='black', shrink=0.05))
plt.plot(f1f2[0], f1f2[1], color = '#ff14f8', marker='.')
plt.annotate('The intersection point between f[1] and f[2]', fontsize = 7, xy=(f1f2[0], f1f2[1]), xytext=(f1f2[0]-5, f1f2[1]-8),
            arrowprops=dict(width = 0.5, headlength = 5, headwidth = 3, facecolor='black', shrink=0.05))
plt.plot(f2f0[0], f2f0[1], color = '#ff14f8', marker='.')
plt.annotate('The intersection point between f[2] and f[0]', fontsize = 7, xy=(f2f0[0], f2f0[1]), xytext=(f2f0[0]-7, f2f0[1]+2),
            arrowprops=dict(width = 0.5, headlength = 5, headwidth = 3, facecolor='black', shrink=0.05))
fillarayX = [f0f1[0], f1f2[0], f2f0[0]]
fillarayY = [f0f1[1], f1f2[1], f2f0[1]]
plt.fill(fillarayX, fillarayY, color='yellow', alpha=0.5)
plt.title('All lines in one figure', fontsize=15)
plt.xlabel('x1', fontsize=12)
plt.ylabel('x2', fontsize=12, color='black')
plt.savefig('my_plot.jpg')
plt.savefig('my_plot.png')
plt.savefig('my_plot.svg')
```

my_plot.png - артефактів майже нема, але обмеження піксельної графіки стають видними при наближенні.

my_plot.svg - векторне зображення, тому якість взагалі не падає при наближенні.
